## Preprocessing Pipeline on X

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
# Numerical_Pipeline
num_pipe = Pipeline(steps=[('impute',SimpleImputer(strategy='mean')),
                          ('scaler',StandardScaler())])
# Categorical_Pipeline
cat_pipe = Pipeline(steps=[('impute',SimpleImputer(strategy='most_frequent')),
                          ('ohe',OneHotEncoder(handle_unknown='ignore'))])
# Combine with column transformer
pre = ColumnTransformer([('num',num_pipe,con),
                        ('cat',cat_pipe,cat)])
```

```python
x_pre = pre.fit_transform(X).toarray()
x_pre
```

```python
cols = pre.get_feature_names_out()
cols
```

```python
x_pre = pd.DataFrame(x_pre,columns=cols)
x_pre
```

## Train, Test Split 

Train - 80% Test - 20%(unseen Data to model)

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x_pre,Y,test_size=0.2,random_state=10)
```

## Tuning the alpha value for lasso

```python
import numpy as np
alphas = np.arange(0.1, 100, 0.1)
alphas
```

```python
from sklearn.metrics import mean_squared_error
tr = []
ts = []
for i in alphas:
    # Train The model
    ls = Lasso(alpha=i)
    ls.fit(xtrain, ytrain)
    # predict the train and test results
    ypred_tr = ls.predict(xtrain)
    ypred_ts = ls.predict(xtest)
    # Calculate MSE in train and test
    mse_tr = mean_squared_error(ytrain, ypred_tr)
    mse_ts = mean_squared_error(ytest, ypred_ts)
    # append values in tr and ts
    tr.append(mse_tr)
    ts.append(mse_ts)
```

```python
tr[0:5]
```

```python
ts[0:5]
```

# Visualize train and test result

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16, 8))
plt.plot(alphas, tr, label='Train MSE')
plt.plot(alphas, ts, label='Test MSE')
plt.xlabel('Alphas')
plt.ylabel('MSE')
plt.legend()
plt.show()
```

# Getting Optimal alpha value

```python
for i in alphas:
    # Train The model
    ls = Lasso(alpha=i)
    ls.fit(xtrain, ytrain)
    # predict the train and test results
    ypred_tr = ls.predict(xtrain)
    ypred_ts = ls.predict(xtest)
    # Calculate MSE in train and test
    mse_tr = mean_squared_error(ytrain, ypred_tr)
    mse_ts = mean_squared_error(ytest, ypred_ts)
    # check the optimum alpha
    if mse_tr>=mse_ts:
        opt_alpha=i
        print(f'Optimum value of alpha : {opt_alpha}')
        break
```

```python
opt_alpha
```

# Create the best lasso model

```python
best_lasso = Lasso(alpha=opt_alpha)
best_lasso.fit(xtrain, ytrain)
```

```python
best_lasso.score(xtrain, ytrain)
```

```python
best_lasso.score(xtest, ytest)
```

# Evaluate the best_lasso model

```python
evaluate_model(xtrain, ytrain, xtest, ytest, best_lasso)
```
